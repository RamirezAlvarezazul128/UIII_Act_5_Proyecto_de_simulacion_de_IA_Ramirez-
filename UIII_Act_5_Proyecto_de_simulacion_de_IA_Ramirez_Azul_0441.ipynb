```python
# --- SISTEMA DE RECOMENDACIÓN DE CANTANTES (IA BÁSICA) ---

print("Bienvenido al Asistente Virtual de Música.")
print("Responde las siguientes preguntas para encontrar tu cantante ideal.\n")

# Entrada de datos: Captura de preferencias del usuario [4-6]
genero = input("¿Qué género musical prefieres? (Pop / Rock / Reggaeton): ").lower()
idioma = input("¿Prefieres música en 'Español' o 'Inglés'?: ").lower()
anio_referencia = int(input("¿Prefieres artistas actuales o clásicos? Ingresa un año de inicio de carrera (ej. 2010): "))

# Lógica de Inteligencia Artificial mediante condicionales [1, 7, 8]
# Se utilizan operadores lógicos 'and' para combinar criterios [9, 10]
# Se utilizan operadores relacionales '==' y '>=' para comparar valores [10, 11]

if genero == "pop" and idioma == "español":
    # Estructura de decisión anidada para filtrar por época [12, 13]
    if anio_referencia >= 2015:
        print("\nIA: Según tu perfil, te recomiendo a Rosalía. Su mezcla de pop y flamenco es tendencia actual.") # [4, 14]
    else:
        print("\nIA: Te recomiendo a Shakira. Sus éxitos de décadas pasadas son pilares del pop latino.") # [15]

elif genero == "pop" and idioma == "inglés":
    if anio_referencia >= 2010:
        print("\nIA: Tu artista ideal es Taylor Swift. Ha dominado las listas de éxitos en los últimos años.") # [4]
    else:
        print("\nIA: Te recomiendo a Madonna. Es la reina del pop con una trayectoria clásica inigualable.") # [15, 16]

elif genero == "rock":
    if idioma == "español":
        print("\nIA: Te recomiendo a Gustavo Cerati. Es una figura legendaria del rock en español.") # [16]
    else:
        # Uso de operadores para decidir entre íconos del rock [17, 18]
        print("\nIA: Te recomiendo a Freddie Mercury. Su voz marcó la historia del rock mundial.") # [1, 16]

elif genero == "reggaeton":
    if anio_referencia >= 2017:
        print("\nIA: Te recomiendo a Bad Bunny. Es el exponente más influyente de la era actual.") # [4]
    else:
        print("\nIA: Te recomiendo a Daddy Yankee. Es considerado el pionero y 'Big Boss' del género.") # [14]

else:
    # Caso por defecto si las preferencias no coinciden con la base de datos [4, 18, 19]
    print("\nIA: Lo siento, aún no tengo un cantante en mi base de datos que se ajuste exactamente a ese perfil.") # [4, 20]

print("\n--- Gracias por usar el recomendador IA ---")
```
